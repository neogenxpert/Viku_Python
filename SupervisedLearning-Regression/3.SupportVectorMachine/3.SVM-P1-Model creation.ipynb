```python
import pandas as pd
```

```python
ds = pd.read_csv("50_Startups.csv")
```

```python
ds=pd.get_dummies(ds,drop_first=True)
```

```python
ds.columns
```

```python
independent = ds[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
dependent = ds [['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=False)
```

```python
x_train
```

```python
## as r2 score is poor we are going to use Standardization to check the model improvement
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
x_train=std.fit_transform(x_train) #fit will claculate the mean and std value
x_test=std.transform(x_test) # here we need of fit as it will take the same above std and mean value.
```

```python
x_train
```

```python
# first two lis will be the procedure to create the model
from sklearn.svm import SVR
regressor = SVR(kernel="linear",C=10) #Kernal = Linear/rbf(non linear)/poly/sigmoid for output. 
#Precomputed wont work as bez it support square matrix only eg 5*5, C penalty can also be used.
# Creats the model
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.support_
```

```python
regressor.n_support_  # How many support vectors have been considered
```

```python
y_pred = regressor.predict(x_test)
```

```python
#calculate the r2 score
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score # is poor and so we are going to try standardization
```
