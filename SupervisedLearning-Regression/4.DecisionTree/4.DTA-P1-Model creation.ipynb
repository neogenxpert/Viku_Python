```python
import pandas as pd
ds = pd.read_csv("50_Startups.csv")
```

```python
ds = pd.get_dummies(ds,drop_first=True)
```

```python
independent = ds[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
dependent = ds [['Profit']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=False)
```

```python
# Decision tree Model creation
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion = 'poisson',splitter ='random')#squared_error”, “friedman_mse”, “absolute_error” (93%), “poisson”(91%)
regressor = regressor.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt # responsible for graphical purpose
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
score
```
