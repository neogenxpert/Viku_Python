```python
import pandas as pd
ds=pd.read_csv("insurance_pre.csv")
```

```python
ds=pd.get_dummies(ds,drop_first=True)
```

```python
independent = ds[["age","bmi","children","sex_male","smoker_yes"]]
dependent= ds[["charges"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Decision tree Model creation
from sklearn.tree import DecisionTreeRegressor
Regressor = DecisionTreeRegressor(criterion = 'poisson',splitter ='random')#squared_error”, “friedman_mse”, “absolute_error” (93%), “poisson”(91%)
Regressor = Regressor.fit(x_train,y_train)
```

```python
#import matplotlib.pyplot as plt # responsible for graphical purpose
#from sklearn import tree
#tree.plot_tree(Regressor)
#plt.show()
```

```python
# Model Learning
y_pred=Regressor.predict(x_test)
```

```python
#R2Score calculation
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
