```python
import pandas as pd
```

```python
ds = pd.read_csv("Social_Network_Ads.csv")
ds
```

```python
ds = pd.get_dummies(ds,drop_first=True)
```

```python
ds['Purchased'].value_counts() # to check balance or inbalanced data set - inbalanced 
```

```python
independent = ds[['Age', 'EstimatedSalary','Gender_Male']]
dependent = ds[['Purchased']]
```

```python
independent.shape # No of records
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
Classifier = RandomForestClassifier(n_estimators= 10,criterion = 'entropy',random_state=0)
#regressor = RandomForestRegressor(n_estimators= 50,criterion='squared_error', random_state=0)
# Estimators = no of tress in forest, default 100.
Classifier.fit(x_train,y_train)
```

```python
y_pred = Classifier.predict(x_test)
```

```python
# Built the matrix
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y_test,y_pred)
print(matrix)
```

```python
# fetch the Classification report
from sklearn.metrics import classification_report
report = classification_report(y_test,y_pred)
print(report)
```
